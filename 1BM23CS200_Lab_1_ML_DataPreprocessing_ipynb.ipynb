```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Select only numerical columns for box plots from the 'data' DataFrame
numerical_cols_data = data.select_dtypes(include=np.number).columns

# Generate box plots for each numerical column in 'data'
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols_data):
    plt.subplot(3, 3, i + 1) # Adjust subplot grid as needed based on the number of numerical columns
    sns.boxplot(y=data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['income'] = le.fit_transform(data['income'])
```

```python
X = data.drop('income', axis=1)
y = data['income']
```

```python
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
X_standardized = X.copy()

X_standardized[numeric_cols] = std_scaler.fit_transform(
    X_standardized[numeric_cols]
)
```

```python
from sklearn.preprocessing import MinMaxScaler

mm_scaler = MinMaxScaler()
X_minmax = X.copy()

X_minmax[numeric_cols] = mm_scaler.fit_transform(
    X_minmax[numeric_cols]
)
```
